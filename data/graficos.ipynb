# Load Data from CSV
Load the data from the CSV file using pandas.

```python
import os

# Check the current working directory
print("Current Working Directory:", os.getcwd())

# List files in the current working directory
print("Files in the current directory:", os.listdir())

# Load the CSV file
import pandas as pd

# Load the CSV file from the 'Trabajo_Algoritmia' directory
data = pd.read_csv('Trabajo_Algoritmia/data/aapl_2014_2023.csv')
```

# Extract First Four Rows
Extract the first four rows of data from the DataFrame.

```python
# Extract the first four rows of data from the DataFrame
rows = data.head(2416)

# Filter rows based on the value in the last column (target)
target_value = 'bearish'  # or 'bullish', depending on what you want to filter
filtered_rows = data[data['target'] == target_value]

filtered_rows
```

# Plot Data
Use matplotlib to create a graph of the first four rows of data.

```python
# Plot Data
plt.figure(figsize=(10, 6))

# Plot 'open' prices
#plt.plot(first_four_rows['date'], first_four_rows['open'], marker='o', label='Open')

# Plot 'high' prices
#plt.plot(first_four_rows['date'], first_four_rows['high'], marker='o', label='High')

# Plot 'low' prices
#plt.plot(first_four_rows['date'], first_four_rows['low'], marker='o', label='Low')
# Plot 'close' prices
plt.plot(filtered_rows['date'], filtered_rows['close'], marker='o', label='Close')

# Add title and labels
plt.title('Valores de cierre en los dias con incremento')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.grid(True)

# Rotate date labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```
